# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```




*   numpy, matplotlib, dan pandas: Digunakan untuk manipulasi data, visualisasi, dan analisis.
*   scipy.cluster.hierarchy & sklearn.cluster: Untuk metode klasterisasi.



## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values  # Columns 3 (Income) and 4 (Spending Score)
```

X = dataset.iloc[:, [3, 4]].values: Mengambil kolom ke-3 (Annual Income) dan ke-4 (Spending Score).

## Using the dendrogram to find the optimal number of clusters

```python
# Generate dendrogram to find the optimal number of clusters
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```






*   dendrogram = sch.dendrogram(sch.linkage(X, method='ward')):
*   sch.linkage: Menghitung jarak antar titik menggunakan metode Ward, yang meminimalkan varian antar cluster.
*   sch.dendrogram: Membuat visualisasi untuk menentukan jumlah cluster optimal (dilihat dari "elbow point").
*   plt.xlabel dan plt.ylabel: Memberi label sumbu agar plot lebih informatif.



## Training the Hierarchical Clustering model on the dataset

```python
# Apply Agglomerative Clustering with the optimal number of clusters (5 as determined visually)
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```







*   c = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward'):
n_clusters: Jumlah cluster yang dihasilkan.
*   metric: Metode perhitungan jarak (Euclidean).
*   linkage: Algoritma pengelompokan (Ward linkage).
*   y_hc = hc.fit_predict(X):
*   fit_predict: Memproses data dan memberikan label cluster untuk setiap data.






## Visualising the clusters

```python
# Plot the clusters
plt.figure(figsize=(10, 7))
colors = ['red', 'blue', 'green', 'cyan', 'magenta']
for i in range(5):  # Adjust range if cluster count changes
    plt.scatter(X[y_hc == i, 0], X[y_hc == i, 1], s=100, c=colors[i], label=f'Cluster {i + 1}')

# Label and display the cluster plot
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```





*   plt.scatter:
Membuat scatter plot untuk setiap cluster dengan warna berbeda.
*  for i in range(5):
Menggunakan loop untuk membuat kode lebih fleksibel jika jumlah cluster berubah.
*   plt.title, plt.xlabel, plt.ylabel, plt.legend, plt.grid:
Memberikan label, legenda, dan grid agar visualisasi lebih informatif.
