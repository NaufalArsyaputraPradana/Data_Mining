### Mencari pencilan dengan Tukey's fences (1)
---

```python
q1 = data_noid.quantile(0.25, numeric_only=True)
q3 = data_noid.quantile(0.75, numeric_only=True)
iqr = q3 - q1
iqr
```

### Mencari pencilan dengan Tukey's fences (2)
---

```python
import warnings
warnings.filterwarnings('ignore')

data_noid_align, iqr_new = data_noid.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (data_noid_align < q1 - 1.5 * iqr_new) | (data_noid_align > q3 + 1.5 * iqr_new)
outlier_filter
```

### Mencari pencilan dengan Tukey's fences (3)
---

```python
data_noid[outlier_filter['assists']] \
    .loc[:, ['player_name', 'assists']] \
    .sort_values(by=['assists'], ascending=False)
```

```python
for col in data_noid.loc[:, 'games':].columns:
    if data_noid[col].dtype == object:
        try:
            data_noid[col] = pd.to_numeric(data_noid[col], errors='raise')
        except ValueError:
            print(f"Column '{col}' contains non-numeric values and cannot be directly converted.")

correlation_matrix = data_noid.loc[:, 'games':].select_dtypes(include=np.number).corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.hist(data_noid['goals'], bins=10)
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.title('Histogram of Goals')
plt.show()

# Scatter Plot
plt.scatter(data_noid['goals'], data_noid['assists'])
plt.xlabel('Goals')
plt.ylabel('Assists')
plt.title('Goals vs Assists')
plt.show()

# Box Plot
sns.boxplot(x=data_noid['goals'])
plt.title('Box Plot of Goals')
plt.show()

# Heatmap Korelasi
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
