Buku tertentu hanya memiliki satu tanggal publikasi. Oleh karena itu perlu dilakukan hal berikut:

    1. Hilangkan tanggal lain dalam kurung siku, 1879[1878]
    2. Konversi rentang tanggal ke “start date”, 1860-63; 1839, 38-54
    3. Hilangkan tanggal yang tidak jelas dan gantikan dengan NaN NumPy, [1879?] -> NaN
    4. Konversi string nan ke nilai NaN  NumPy

```python
# Fungsi untuk membersihkan kolom 'Date of Publication'
def clean_date(item):
    if isinstance(item, str):
        item = item.strip()
        if '[' in item:
            item = item.split('[')[0].strip()
        if '-' in item:
            item = item.split('-')[0].strip()
        if ',' in item:
            item = item.split(',')[0].strip()
        if '?' in item:
            return np.nan
        return item
    return item


# Bersihkan kolom 'Date of Publication'
df['Date of Publication'] = df['Date of Publication'].apply(clean_date)


# Konversi string 'nan' ke nilai NaN
df['Date of Publication'] = df['Date of Publication'].replace('nan', np.nan)

# Periksa hasil
print(df['Date of Publication'].head(20))
```

```python
# menggunakan ekspresi reguler (Regex) tunggal untuk mengekstrak tahun publikasi

regex = r'^(\d{4})'
```

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].isnull().sum() / len(df)
```

```python
def get_citystate(item):
     if ' (' in item:
         return item[:item.find(' (')]
     elif '[' in item:
         return item[:item.find('[')]
     else:
         return item
```

```python
towns_df =  towns_df.applymap(get_citystate)
```
